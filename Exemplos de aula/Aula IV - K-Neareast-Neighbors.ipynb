```python
from sklearn import neighbors
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt

x = [[2.0, 1.5, 1.0], [1.0, 1.0, 0.5], [1.0, 3.0, 1.5], [0.8, 2.2, 0.7], [3.0, 1.0, 0.2], [0.4, 2.0, 1.7], [1.2, 1.2, 1.5]]
y = [1,0,1,0,1,1,1]

clf = neighbors.KNeighborsClassifier(n_neighbors = 3)
clf = clf.fit(x,y)

print(clf.predict([[0.9, 1.1, 0.5]])) ## Predict possibilita de ver a reação do classificador de acordo com múltiplas entradas.
print(clf.predict_proba([[0.9, 1.1, 0.5]]))
```

```python
x = [[2.0, 1.5], [1.0, 1.0], [1.0, 3.0], [0.8, 2.2], [3.0, 1.0], [0.4, 2.0], [1.2, 1.2]]
y = [1,0,1,0,1,1,1]

X = np.array(x)
Y = np.array(y)

clf = neighbors.KNeighborsClassifier(n_neighbors = 3)
clf = clf.fit(X,Y)

step_size = 0.01
x_min, x_max = X[:, 0].min(), X[:, 0].max()
y_min, y_max = X[:, 1].min(), X[:, 1].max()

x_min, x_max = x_min-abs(0.1*x_min), x_max+abs(0.1*x_max) 
y_min, y_max = y_min-abs(0.1*y_min), y_max+abs(0.1*y_max) 

xx, yy = np.meshgrid(np.arange(x_min, x_max, step_size), np.arange(y_min, y_max, step_size))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure()
plt.pcolormesh(xx, yy, Z, cmap='Set3')

plt.scatter(X[:,0], X[:,1], c=Y, edgecolor = 'k', s=20, cmap = 'Set3')
plt.show()

```

```python
import matplotlib.animation as animation
dataset = datasets.load_breast_cancer()

X = dataset.data[:, 0:2]
Y = dataset.target

k = 1
imgs = []

fig = plt.figure()
ax = fig.add_subplot(111)

while(k<569+1):
    clf = neighbors.KNeighborsClassifier(n_neighbors = k)
    clf = clf.fit(X,Y)
    
    step_size = 0.01
    x_min, x_max = X[:, 0].min(), X[:, 0].max()
    y_min, y_max = X[:, 1].min(), X[:, 1].max()

    x_min, x_max = x_min-abs(0.1*x_min), x_max+abs(0.1*x_max) 
    y_min, y_max = y_min-abs(0.1*y_min), y_max+abs(0.1*y_max) 

    xx, yy = np.meshgrid(np.arange(x_min, x_max, step_size), np.arange(y_min, y_max, step_size))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    imgplot1 = plt.pcolormesh(xx, yy, Z, cmap='Set3')
    imgplot2 = plt.scatter(X[:,0], X[:,1], c=Y, edgecolor = 'k', s=20, cmap = 'Set3')
    titlestr = "k = " + str(k)
    ttl = plt.text(0.5, 1.01, titlestr, horizontalalignment ='center', verticalalignment='bottom', transform = ax.transAxes)
    imgs.append([imgplot1, imgplot2,ttl])
    k = k+2
    
im_ani = animation.ArtistAnimation(fig, imgs, interval = 500, repeat = False, blit= False)
im_ani.save('im.mp4')
    
```

```python
X = np.random.random([1000,2])
Y = []
for x_value in X:
    y_value = 0
    if(((x_value[0] <= 0.5) and (x_value[1] > 0.5)) or ((x_value[1] <= 0.5) and (x_value[0] > 0.5))):
        y_value = 1
    
    Y.append(y_value)

X = np.array(X)
Y = np.array(Y)

X[:,1] = 100*X[:,1]

clf = neighbors.KNeighborsClassifier(n_neighbors=3)
clf = clf.fit(X,Y)

step_size = 0.01
x_min, x_max = X[:, 0].min(), X[:, 0].max()
y_min, y_max = X[:, 1].min(), X[:, 1].max()

xx, yy = np.meshgrid(np.arange(x_min, x_max, step_size), np.arange(y_min, y_max, step_size))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure()
plt.pcolormesh(xx, yy, Z, cmap='Set3')

plt.scatter(X[:,0], X[:,1], c=Y, edgecolor = 'k', s=20, cmap = 'Set3')
plt.show()

```
