```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split ## Importando biblioteca que separa o dataset de treino e de teste aleatoriamente e automática
from sklearn import tree
from sklearn.metrics import accuracy_score
import numpy as np
import graphviz
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'

data = load_breast_cancer()

label_names = data['target_names']
labels = data['target']
feature_names = data['feature_names']
features = data['data']

train, valid, train_labels, valid_labels = train_test_split(features, labels, test_size = 0.4, random_state=42)
validr, test, validr_labels, test_labels = train_test_split(valid, valid_labels, test_size = 0.5, random_state=42)


j=[]
for i in range (2, 101):
    arv = tree.DecisionTreeClassifier(max_depth = i)
    arv = arv.fit(train, train_labels)
    predicted_results = arv.predict(validr)
    results = accuracy_score(predicted_results, validr_labels)
    j.append(results)
    
j = np.array(j)
idx = np.argmax(j) + 2 ## Putting the index value into the same range of the parameters used on the max_depth input
print("A profundidade escolhida é a:", idx)

treinamento = np.concatenate((train, validr))
treinamento_labels = np.concatenate((train_labels, validr_labels))

arvt = tree.DecisionTreeClassifier(max_depth = idx)
arvt = arvt.fit(treinamento, treinamento_labels)
predicted_test = arvt.predict(test)
final_results = accuracy_score(predicted_test, test_labels)*100
print("A acuraciedade do nosso modelo ficou de", final_results, "%")

n = []
for l in range (2, 101):
    for m in range (0, 10):
        arvt = tree.DecisionTreeClassifier(max_depth = l)
        arvt = arvt.fit(treinamento, treinamento_labels)
        predicted_test = arvt.predict(test)
        final_results = ((accuracy_score(test_labels, predicted_test)*100)-l)
        n.append(final_results)
   
n = np.array(n)
print(n.spape)
print(n)
##n = np.mean(n)

##print("O desempenho corrigido pela maior quantidade de profundidade é:", n, "%")


```
