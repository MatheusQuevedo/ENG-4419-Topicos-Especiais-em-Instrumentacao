```python
from interval import interval, inf, imath

a = interval[2,5]
b = interval[1,10]
c = interval[-10,30]
d = interval[7,10]

e = a+b
f = a-b
g = a*c
h = a/b
t = a/c



print("A uniao de A com B é:", (a | b))
print("O casco da uniao de A com B é:", interval.hull([a,b]))
print("A uniao de A com D é:", (a | d))
print("O casco da uniao de A com D é:", interval.hull([a,d]))
print("A interseccao de A com B é:", (a & b))
print("A interseccao de A com D é:", (a & d))
print("A adicao de A com B é:", e)
print("A subtracao de A com B é:", f)
print("A multiplicacao de A com c é:", g)
print("A divisao de A com B é:", h)
print("A divisao de A com c é:", t)
```
