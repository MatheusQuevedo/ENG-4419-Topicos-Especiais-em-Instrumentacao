```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.model_selection import cross_val_score
```

```python
data = pd.read_csv("abalone.data", names = ["1", '2', '3', '4', '5', '6', '7', '8', '9'])
targ = data.values.tolist()

## working the dataset
tr = data.drop(columns = ['9'])
tr["1"] = tr['1'].astype('category')
tr['1_cat'] = tr['1'].cat.codes
tr = tr.drop(columns = ['1'])
tr = tr.values.tolist()
#print(tr)
inputs=[]
for i in targ:
    inputs.append(i[8])

    
train, valid, train_labels, valid_labels = train_test_split(tr, inputs, test_size = 0.33)


arv = tree.DecisionTreeClassifier()
resultado = (cross_val_score(arv,train,train_labels,cv=10)*100)
end = np.mean(resultado)
print(end)


j=[]
for i in range (2, 101):
        arv = tree.DecisionTreeClassifier(max_depth = i)
        arv = arv.fit(train, train_labels)
        resultado = (np.mean((cross_val_score(arv,train,train_labels,cv=10)*100)))
        #resultado = (((cross_val_score(arv,train,train_labels,cv=10)*100)))
        j.append(resultado)
    
j = np.array(j)
print(j)
idx = (np.argmax(j)) + 2
print("A profundidade escolhida é a:", idx)




```

```python
j=[]
for i in range (1, 101):
        arv = tree.DecisionTreeClassifier(max_depth = 4, min_samples_leaf=i)
        arv = arv.fit(train, train_labels)
        resultado = (np.mean(cross_val_score(arv,train,train_labels,cv=10)*100))
        j.append(resultado)
    
j = np.array(j)
print(j)
idx = (np.argmax(j)) + 1
print("A min_samples_leaf é a:", idx)
```

```python
j=[]
for i in range (2, 101):
        arv = tree.DecisionTreeClassifier(max_depth = 4, min_samples_leaf=47, max_leaf_nodes = i)
        arv = arv.fit(train, train_labels)
        resultado = (np.mean(cross_val_score(arv,train,train_labels,cv=10)*100))
        j.append(resultado)
    
j = np.array(j)
print(j)
idx = (np.argmax(j)) + 2
print("O max_leaf_nodes escolhida é a:", idx)
```

```python
arv = tree.DecisionTreeClassifier(max_depth = 4, min_samples_leaf=47, max_leaf_nodes = 14)
##arv = arv.fit(train, train_labels)
resulta = cross_val_score(arv,train,train_labels,cv=10)
print("Resultado evitando o overfitting:", np.mean(resulta))
```

```python
from sklearn.tree import DecisionTreeClassifier

arv = DecisionTreeClassifier()
path = arv.cost_complexity_pruning_path(train, train_labels)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

clfs = []
for ccp_alpha in ccp_alphas:
    arvs = tree.DecisionTreeClassifier(random_state=0, ccp_alpha = ccp_alpha)
    arvs.fit(train, train_labels)
    clfs.append(arvs)

clfs = np.array(clfs)    
train_scores = [np.mean((cross_val_score(arv,train,train_labels,cv=10))) for arvs in clfs]
idx = (np.argmax(clfs))
print("A melhor arvore é a do index:", idx)
result = clfs[idx]
print("O resultado do score é:", result)
```

```python
from sklearn.ensemble import BaggingClassifier

j=[]
for i in range (1,100):
    arv = BaggingClassifier(base_estimator = tree.DecisionTreeClassifier(), n_estimators = i, random_state = 0, oob_score = True)
    arv.fit(train, train_labels)
    resultado = arv.oob_score_
    j.append(resultado)

j = np.array(j)
print(j)
idx = (np.argmax(j))
print("A melhor arvore é a do index:", idx)
print("Resultado", j[idx])
```

```python
from sklearn.ensemble import BaggingClassifier

j=[]
for i in range (1,100):
    arv = BaggingClassifier(base_estimator = tree.DecisionTreeClassifier(), n_estimators = i, random_state = 0)
    arv.fit(train, train_labels)
    resultado = (np.mean(cross_val_score(arv,train,train_labels,cv=10)*100))
    j.append(resultado)

j = np.array(j)
idx = (np.argmax(j))
print(j[idx])
print("A melhor arvore é a do index:", idx)
```

```python
from sklearn.ensemble import RandomForestClassifier

j=[]
for i in range(1, 100):
    arv = RandomForestClassifier(n_estimators=i, max_features = "sqrt", random_state = 0)
    arv.fit(train, train_labels)
    resultado = (np.mean(cross_val_score(arv,train,train_labels,cv=10)*100))
    j.append(resultado)

j = np.array(j)
idx = (np.argmax(j))
print(j[idx])
print("A melhor arvore é a do index:", idx)
```
